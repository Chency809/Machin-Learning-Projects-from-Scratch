## Read Text Corpus

```python
import docx
import numpy as np
import tensorflow as tf
from tqdm import tqdm
import pickle
```

```python
doc = docx.Document('诗.docx')
```

```python
data = ""
all_sentences = []
for para in doc.paragraphs:
    text = para.text.replace('    ', '，').replace(' ', '，').replace('，，', '，').replace('，，', '，').replace('！', '。').replace('，', '。')
    text = text.replace('。', '\n')
    
    if text[-1] == '\n':
        text = text[:-1]
    
    text = text.split('\n')
    all_sentences += text
    data = '\n'.join(all_sentences)

print(data)
```

## Skip-gram Train Data Extraction

```python
all_int_sentences = np.zeros((725, 9))
X = np.zeros((725 * 10, 2))
y = np.zeros((725 * 10, 1))
k = 9

for i, s in enumerate(tqdm(all_sentences)):
    int_s = [vocab_stoi[i] for i in list(s)]
    
    if len(int_s) != 1:
        context = np.random.randint(0, len(int_s))
        pos_target = np.random.randint(0, len(int_s))
        
        while context == pos_target:
            pos_target = np.random.randint(0, len(int_s))
        else:
            X[i * 10] = np.array([int_s[context], int_s[pos_target]]).reshape(1, 2)
            y[i * 10] = 1
            
        for n in range(k):
            neg_target = np.random.randint(0, 997)
            X[n + i * 10 + 1] = np.array([int_s[context], neg_target]).reshape(1, 2)
        
    while len(int_s) != 9:
        int_s.append(0)
            
    all_int_sentences[i] = np.array(int_s).reshape(1, 9) 
```

## Word Embeddings Training

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from functools import partial

batch_size = 50
buffer_size = 1000

train_data = tf.data.Dataset.from_tensor_slices((X, y))
train_data = train_data.cache().shuffle(buffer_size).batch(batch_size).repeat()

print(X.shape)
print(train_data)

embedding_dim = 16
vocab_size = 997

model = Sequential([
    Embedding(vocab_size, embedding_dim),
    LSTM(64, return_sequences=False),
    Dense(16, activation='tanh'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001, 
                                             beta_1=0.999, beta_2=0.999, epsilon=10**-8),
                                             loss='binary_crossentropy',
                                             metrics=['accuracy'])
model.summary()
```

```python
train_step = len(X) // batch_size
```

```python
model.fit(x=train_data, 
          epochs=100,
          steps_per_epoch=train_step,
          verbose=2)
```

```python
e = model.layers[0]
weights = e.get_weights()[0]
print(weights.shape) # shape: (vocab_size, embedding_dim)
```

## Model building

```python
BATCH_SIZE = 1
BUFFER_SIZE = 725

train_data = tf.data.Dataset.from_tensor_slices((sentences, y))
train_data = train_data.shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True).repeat()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout
from tensorflow.keras.initializers import Constant

model = tf.keras.models.Sequential([
    Embedding(word_matrix.shape[0], word_matrix.shape[1], 
              batch_input_shape=[BATCH_SIZE, None], # None is compatible with any sizes
              embeddings_initializer=Constant(word_matrix),
              trainable=False),
    LSTM(256, return_sequences=True, stateful=True), # stateful requires to state batch size
    Dense(1024, activation='tanh'),
    Dense(word_matrix.shape[0], activation='tanh'),
    Dense(word_matrix.shape[0], activation='softmax'),
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001, 
                                                 beta_1=0.99, beta_2=0.999, epsilon=10**-8),
                                                 loss='categorical_crossentropy')

model.summary()
```

```python
import os 

checkpoint_dir = 'training_checkpoints/'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt_{epoch}")

checkpoint_callback=tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True,
    save_freq=5)
```

```python
model.fit(x=train_data, 
          epochs=50,
          steps_per_epoch=len(sentences) // BATCH_SIZE,
          callbacks=[checkpoint_callback],
          verbose=2)
```

```python
def sample(preds, vocab_size, itos):
    idx = np.random.choice(np.arange(vocab_size), 1, True, preds)[0]
    
    return idx
```

```python
def generate_text_xjh(model, num_generate, num_per_row):
    # Evaluation step (generating text using the learned model)

    # Converting our start string to numbers (vectorizing)
    input_eval = np.zeros([1])
    input_eval = tf.expand_dims(input_eval, 0)

    # Empty string to store our results
    text_generated = ''
    
    row = 0
    word = 0
    predicted_id = None
    for i in range(num_generate):
        
        if i != 0 and i % num_per_row == 0:
            text_generated += '\n'
            row += 1
            word = 0
            
        pause_cond = ((row == 0)and(word == 6)) or ((row == 2)and(word == 3 or word == 6))
        
        if pause_cond:
            text_generated += ','
            
        predictions = model2.predict(input_eval)
        # remove the batch dimension
        predictions =  tf.squeeze(predictions, [0, 1]).numpy()

        # using a categorical distribution to predict the character returned by the model
        predicted_id = sample(predictions, 997, itos)
        
        while predicted_id == stoi['\n']:
            predicted_id = sample(predictions, 997, itos)
        # We pass the predicted character as the next input to the model
        # along with the previous hidden state
        input_eval = tf.expand_dims([predicted_id], 0)

        text_generated += itos[predicted_id]
        word += 1
        
    return (text_generated)
```

```python
def generate_text(model, num_generate, num_per_row):
    # Evaluation step (generating text using the learned model)

    # Converting our start string to numbers (vectorizing)
    input_eval = np.zeros([1])
    input_eval = tf.expand_dims(input_eval, 0)

    # Empty string to store our results
    text_generated = ''
    
    count = 0
    predicted_id = None
    
    for i in range(num_generate):
        
        if i!= 0 and i % num_per_row == 0:
            text_generated += '\n'
        
        predictions = model2.predict(input_eval)
        # remove the batch dimension
        predictions =  tf.squeeze(predictions, [0, 1]).numpy()

        # using a categorical distribution to predict the character returned by the model
        predicted_id = sample(predictions, 997, itos)
        
        while predicted_id == stoi['\n']:
            predicted_id = sample(predictions, 997, itos)
        # We pass the predicted character as the next input to the model
        # along with the previous hidden state
        input_eval = tf.expand_dims([predicted_id], 0)

        text_generated += itos[predicted_id]
        
    return (text_generated)
```

```python
def cosine_similarity(u, v):
    """
    Cosine similarity reflects the degree of similarity between u and v
        
    Arguments:
        u -- a word vector of shape (n,)          
        v -- a word vector of shape (n,)

    Returns:
        cosine_similarity -- the cosine similarity between u and v defined by the formula above.
    """
    
    distance = 0.0

    dot = np.dot(u, v)
    norm_u = np.sqrt(np.sum(u**2))
    norm_v = np.sqrt(np.sum(v**2))
    cosine_similarity = dot/(norm_u * norm_v)
    
    return cosine_similarity
```
